```python
# removing features which has more null values

rmv_null_col = null_val_p[null_val_p > 17].keys()
rmv_null_col
```

```python
new_df = df.drop(columns=['LotFrontage', 'Alley', 'FireplaceQu', 'PoolQC', 'Fence',
       'MiscFeature'])
new_df.head()
```

```python
new_df.shape
```

```python
print(new_df.isnull().sum().sum())
```

```python
# separating numerical valued features and fill missing values

num_val_col = new_df.select_dtypes(include=['int64', 'float64']).columns
num_val_col
```

```python
num_val_col_df = new_df.select_dtypes(include=['int64', 'float64'])
num_val_col_df.head()
```

```python
# extracting columns which has null numerical values

null_num_val = [var for var in num_val_col if new_df[var].isnull().sum() > 0]
null_num_val
```

```python
# applying data filling strategies
# 1. mean, median method

new_df_median = new_df.fillna(num_val_col_df.median()[0])
print(new_df_median.shape)
```

```python
print(new_df_median['MasVnrArea'].isnull().sum())
print(new_df_median['MasVnrArea'].median())
print(new_df_median['GarageYrBlt'].median())
```

```python
new_df.isnull().sum().sum()
```

```python
# 2. using simple imputer method

from sklearn.impute import SimpleImputer
```

```python
s_imputer = SimpleImputer(strategy='mean')
```

```python
s_imputer.fit(new_df[num_val_col])
```

```python
new_df[num_val_col] = s_imputer.transform(new_df[num_val_col])
new_df[num_val_col].head()
```

```python
new_df[num_val_col].shape
```

```python
new_df[num_val_col].isnull().sum().sum()
```

```python
new_df.isnull().sum().sum()
```

```python
# separating categorical valued features and fill missing values
# 1. using mode method

cat_val_col = new_df.select_dtypes(include='object').columns
cat_val_col
```

```python
cat_val_col_df = new_df.select_dtypes(include='object')
cat_val_col_df.head()
```

```python
cat_val_col_df.shape
```

```python
null_cat_val_col = [var for var in cat_val_col if new_df[var].isnull().sum() > 0]
print(null_cat_val_col, end=' ')
```

```python
for var in null_cat_val_col:
    new_df[var].fillna(new_df[var].mode()[0], inplace = True)
    print(var, "=", new_df[var].mode()[0])
```

```python
print(new_df.shape)
print(new_df.isnull().sum().sum())
```

```python
# converting categorical values into numerical values using label encoding

from sklearn.preprocessing import LabelEncoder, OrdinalEncoder
```

```python
ord_enc = OrdinalEncoder()
```

```python
new_df[cat_val_col] = ord_enc.fit_transform(new_df[cat_val_col])
new_df[cat_val_col].head()
```

```python
up_df = pd.concat([new_df[num_val_col], new_df[cat_val_col]], axis=1)
up_df.head()
```
