```python
# % of null values

null_percent = df.isnull().mean()*100
null_percent
```

### Removing columns which has greater null values (more than 17%)

```python
remove_cols = null_percent[null_percent > 17].keys()
remove_cols
```

```python
new_df = df.drop(columns=remove_cols)
new_df.shape
```

```python
new_df.info()
```

### Selecting Numerical Valued features


```python
int_col_id = new_df.select_dtypes(include=['int64', 'float64']).columns
int_col_id
```

```python
int_col_df = new_df.select_dtypes(include=['int64', 'float64'])
int_col_df.head()
```

```python
int_null_lst = []

for var in int_col_id:
    if new_df[var].isnull().sum() > 0:
        int_null_lst.append(var)

int_null_lst
```

```python
int_null_lst_1 = [var for var in int_col_id if new_df[var].isnull().sum() > 0]
int_null_lst_1
```

```python
new_df_1 = new_df.fillna(int_col_df.mean())
new_df_1.shape
```

```python
new_df_1.isnull().sum()
```

### For categorical values

```python
obj_col_id = new_df.select_dtypes(include='object').columns
obj_col_id
```

```python
obj_col_df = new_df.select_dtypes(include='object')
obj_col_df.head()
```

```python
obj_val_percent = obj_col_df.isnull().mean()*100
obj_val_percent
```

```python
obj_null_val = obj_val_percent[obj_val_percent > 0].keys()
obj_null_val
```

```python
obj_null_val_1 = [var for var in obj_col_id if new_df[var].isnull().sum() > 0]
obj_null_val_1
```

```python
for var in obj_null_val:
    new_df_1[var].fillna(new_df_1[var].mode()[0], inplace = True)
    print(var, "=", new_df_1[var].mode()[0])
```
