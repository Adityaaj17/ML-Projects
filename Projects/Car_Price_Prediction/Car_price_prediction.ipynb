## Car Price Prediction using ML Algorithms

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("P:\Artificial Intelligence\Machine Learning\Projects\Car_Price_Prediction\Car_data.csv")
```

```python
pd.set_option('display.max_columns', None)
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(columns=['Unnamed: 0', 'Location', 'Kilometers_Driven', 'Owner_Type', 'New_Price'], inplace=True)
```

```python
df.info()
```

```python
null_pect = df.isnull().sum()
null_pect
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
for var in df:
    print(df[var].value_counts())
```

```python
df.drop(columns=['Name'], inplace=True)
```

```python
df.head()
```

```python
df2 = df[['Mileage', 'Engine', 'Power']]
```

```python
df2.head()
```

```python
df2['Mileage'] = df2['Mileage'].str.replace('km/kg', 'kmpl')
df2['Mileage'] = df2['Mileage'].str.replace('kmpl', '').astype(float)
df2['Engine'] = df2['Engine'].str.replace('CC', '').astype(float)
df2['Power'] = df2['Power'].str.replace('bhp', '')
df2['Power'] = df2['Power'].replace(['null', 'null '], '0').astype(float)
df2['Power'] = df2['Power'].astype(float)
```

```python
df2.head()
```

```python
df4 = df.drop(columns=['Mileage', 'Engine', 'Power'])
```

```python
df4.head()
```

```python
final_df = pd.concat([df2, df4], axis=1)
```

```python
final_df.head()
```

```python
final_df.info()
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
for var in final_df:
    if final_df[var].dtype == 'object':
        final_df[var] = encoder.fit_transform(final_df[var])
```

```python
final_df.head()
```

```python
from sklearn.model_selection import train_test_split

f_class = final_df.drop(columns='Price')
t_class = final_df['Price']

X_train, X_test, y_train, y_test = train_test_split(f_class, t_class, test_size=0.2, random_state=42)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

sc_scaler = StandardScaler()
X_train_scaled = sc_scaler.fit_transform(X_train)
X_test_scaled = sc_scaler.transform(X_test)
print(X_train_scaled.shape, X_test_scaled.shape)
```

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
fit_model = regressor.fit(X_train_scaled, y_train)
fit_model.score(X_test_scaled, y_test)
```

```python
y_pred = fit_model.predict(X_test_scaled)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

print('Linear Regression')
print('Mean Absolute Error: ', mean_absolute_error(y_test, y_pred))
print('Mean Squarred Error: ', mean_squared_error(y_test, y_pred))
print('Root Mean Squarred Error: ', np.sqrt(mean_squared_error(y_test, y_pred)))
print('Accuracy:', fit_model.score(X_test_scaled, y_test)*100, '\n')
```

```python
from sklearn.ensemble import RandomForestRegressor

rf_regressor = RandomForestRegressor(n_estimators=65, random_state=51)
fit_rfr_model = rf_regressor.fit(X_train, y_train)
fit_rfr_model.score(X_test, y_test)
```

```python
y_pred2 = fit_rfr_model.predict(X_test)
```

```python
print('Ranom Forest Regressor')
print('Mean Absolute Error: ', mean_absolute_error(y_test, y_pred2))
print('Mean Squarred Error: ', mean_squared_error(y_test, y_pred2))
print('Root Mean Squarred Error: ', np.sqrt(mean_squared_error(y_test, y_pred2)))
print('Accuracy:', fit_rfr_model.score(X_test, y_test)*100, '\n')
```
